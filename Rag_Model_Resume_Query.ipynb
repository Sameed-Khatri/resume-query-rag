```python
import PyPDF2

# Open the PDF file
with open('./Resume.pdf', 'rb') as file:
    reader = PyPDF2.PdfReader(file)

    # Extract text from each page
    cv_text = ""
    for page in range(len(reader.pages)):
        cv_text += reader.pages[page].extract_text()

# Display extracted text
cv_text
```

```python
import re

# Replace all types of whitespace (including tabs and newlines) with a single space
cleaned_text = re.sub(r'\s+', ' ', cv_text).strip()
cleaned_text
```

### **TO CREATE CHUNKS BASED ON TAGS IN RESUME**

```python
import re

# Cleaned text from the resume
resume_text = cleaned_text

# Define sections by tags
sections = {
    'Contact Details': r'Phone:',
    'Email': r'E-mail:',
    'LinkedIn': r'LinkedIn:',
    'Objective': r'Objective',
    'Academic Qualification': r'Academic Qualification',
    'Achievements': r'Achievements',
    'Professional Experience': r'Professional Experience',
    'Major Projects': r'Major Reports/ Research Work/Projects',
    'Extracurricular': r'Interests and Extracurricular Activities',
    'Skills': r'Skills'
}

# Function to chunk text by sections
def chunk_by_tags(text, sections):
    chunks = {}
    for tag, pattern in sections.items():
        match = re.search(pattern, text)
        if match:
            start = match.end()
            next_tag_pos = min([re.search(next_pattern, text[start:]).start() + start for next_pattern in sections.values() if re.search(next_pattern, text[start:])], default=len(text))
            chunks[tag] = text[start:next_tag_pos].strip()
    return chunks

# Apply the function to your resume text
chunks = chunk_by_tags(resume_text, sections)

# Output the chunks (you can print or use them for embedding)
for tag, chunk in chunks.items():
    print(f"{tag}:\n{chunk}\n")
```

### **TO CREATE CHUNKS USING RECURSIVE CHARACTER TEXT SPLITTER**

```python
txt_split = RecursiveCharacterTextSplitter(chunk_size = 300, chunk_overlap = 120)
chunk_txt = txt_split.split_text(cleaned_text)
chunk_txt
```

```python
from langchain import embeddings
persist_directory = './vectorDB'

embedding = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

vectorDB = Chroma.from_texts(texts=chunk_txt, embedding=embedding, persist_directory=persist_directory)
```
